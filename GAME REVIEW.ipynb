```python
from google_play_scraper import Sort,reviews
result,continuation_token=reviews(
'com.ludo.king',
lang='en',
country='us',
sort=Sort.NEWEST,
count=1000,
filter_score_with=5)
result,_=reviews('com.ludo.king',
                continuation_token=continuation_token)
```

```python
print(result)
```

```python
import pandas as pd
import numpy as np
```

```python
ludo_king=pd.DataFrame(np.array(result),columns=['review'])
ludo_king=ludo_king.join(pd.DataFrame(ludo_king.pop('review').tolist()))
ludo_king.head()
```

```python
from IPython.display import HTML
import base64
def create_download_link( data, title = "Download CSV file", filename = "ludo_king.csv"):  
    csv = ludo_king.to_csv()
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)
create_download_link(ludo_king)
```
